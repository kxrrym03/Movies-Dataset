```python
# Group data by ReleaseYear and sum the Global_BoxOfficeUSD
yearly_box_office = df.groupby('ReleaseYear')['Global_BoxOfficeUSD'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(yearly_box_office['ReleaseYear'], yearly_box_office['Global_BoxOfficeUSD'], marker='o', linestyle='-', alpha=0.8)

# Labels and title
plt.title('Total Global Box Office Revenue by Year', fontsize=16)
plt.xlabel('Release Year', fontsize=14)
plt.ylabel('Global Box Office (USD)', fontsize=14)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

```python
sns.countplot(data=df, x='Genre')
plt.title('Genre Distribution')
plt.xlabel('Genre')
plt.xticks(rotation=90)
plt.ylabel('Count')
plt.show()
```
